```python
import pennylane as qml

dev = qml.device("default.qubit", wires=8)

@qml.qnode(dev)
def circuit(params):
    qml.RY(params[0], wires=0)
    qml.RY(params[0], wires=1)
    qml.RY(params[0], wires=2)
    qml.RY(params[0], wires=3)
    qml.RY(params[0], wires=4)
    qml.RY(params[0], wires=5)
    qml.RY(params[0], wires=6)
    qml.RY(params[0], wires=7)
    qml.RY(params[0], wires=0)
    qml.RY(params[0], wires=1)
    qml.RY(params[0], wires=2)
    qml.RY(params[0], wires=3)
    qml.RY(params[0], wires=4)
    qml.RY(params[0], wires=5)
    qml.RY(params[0], wires=6)
    qml.RY(params[0], wires=7)
    qml.RY(params[0], wires=0)
    qml.RY(params[0], wires=1)
    qml.RY(params[0], wires=2)
    qml.RY(params[0], wires=3)
    qml.RY(params[0], wires=4)
    qml.RY(params[0], wires=5)
    qml.RY(params[0], wires=6)
    qml.RY(params[0], wires=7)
    qml.RY(params[0], wires=0)
    qml.RY(params[0], wires=1)
    qml.RY(params[0], wires=2)
    qml.RY(params[0], wires=3)
    qml.RY(params[0], wires=4)
    qml.RY(params[0], wires=5)
    qml.RY(params[0], wires=6)
    qml.RY(params[0], wires=7)
    qml.RY(params[0], wires=0)
    qml.RY(params[0], wires=1)
    qml.RY(params[0], wires=2)
    qml.RY(params[0], wires=3)
    qml.RY(params[0], wires=4)
    qml.RY(params[0], wires=5)
    qml.RY(params[0], wires=6)
    qml.RY(params[0], wires=7)
    qml.RY(params[0], wires=0)
    qml.RY(params[0], wires=1)
    qml.RY(params[0], wires=2)
    qml.RY(params[0], wires=3)
    qml.RY(params[0], wires=4)
    qml.RY(params[0], wires=5)
    qml.RY(params[0], wires=6)
    qml.RY(params[0], wires=7)
    qml.RY(params[0], wires=0)
    qml.RY(params[0], wires=1)
    qml.RY(params[0], wires=2)
    qml.RY(params[0], wires=3)
    qml.RY(params[0], wires=4)
    qml.RY(params[0], wires=5)
    qml.RY(params[0], wires=6)
    qml.RY(params[0], wires=7)
    return qml.expval(qml.PauliZ(0)), qml.expval(qml.PauliZ(1)), qml.expval(qml.PauliZ(2)), qml.expval(qml.PauliZ(3)), qml.expval(qml.PauliZ(4)), qml.expval(qml.PauliZ(5)), qml.expval(qml.PauliZ(6)), qml.expval(qml.PauliZ(7))
params =[0]

print(qml.draw_mpl(circuit, style='sketch')(params))
```
