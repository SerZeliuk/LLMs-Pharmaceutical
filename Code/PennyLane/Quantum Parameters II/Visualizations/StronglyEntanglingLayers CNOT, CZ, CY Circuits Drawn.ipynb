```python
import pennylane as qml
import numpy as np

dev = qml.device('default.qubit', wires=4)

@qml.qnode(dev)
def circuit(feature_vector):
    qml.AngleEmbedding(features=feature_vector, wires=range(4))
    return qml.probs(wires=range(4))

X = [1,2,3,4]
print(qml.draw_mpl(circuit, expansion_strategy="device")(X))
```

```python
import pennylane as qml
import numpy as np

dev = qml.device('default.qubit', wires=4)
n_qubits=4

@qml.qnode(dev)
def circuit(parameters):
    qml.StronglyEntanglingLayers(weights=parameters, wires=range(4))
    return qml.expval(qml.PauliZ(0))

# 8 Layers CNOT
shape = qml.StronglyEntanglingLayers.shape(n_layers=8, n_wires=4)
weights = np.random.random(size=shape)
print(qml.draw_mpl(circuit, expansion_strategy="device")(weights))
```
