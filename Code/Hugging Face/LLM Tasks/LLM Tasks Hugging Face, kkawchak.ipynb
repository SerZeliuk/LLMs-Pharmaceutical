```python



# Sentiment Analysis using 10 Different Hugging Face LLMs
# Rankings based on two positive labels with highest scores



# 01 distilbert/distilbert-base-uncased-finetuned-sst-2-english
# 02 RecordedFuture/Swedish-Sentiment-Fear
# 03 siebert/sentiment-roberta-large-english
# 04 SchuylerH/bert-multilingual-go-emtions
# 05 finiteautomata/bertweet-base-sentiment-analysis
# 06 pysentimiento/robertuito-sentiment-analysis
# 07 lxyuan/distilbert-base-multilingual-cased-sentiments-student
# 08 nlptown/bert-base-multilingual-uncased-sentiment
# 09 yiyanghkust/finbert-tone
# 10 StephanAkkerman/FinTwitBERT-sentiment


```

```python
classifier = pipeline("sentiment-analysis", model = "distilbert/distilbert-base-uncased-finetuned-sst-2-english")

# Modify text for sentiment analysis here
results = classifier(["Large Language Models are really awesome!", "Medical Generative AI is very practical!"])
for result in results:
    print(f"label: {result['label']}, with score: {round(result['score'] * 100, 4)}%")
```
