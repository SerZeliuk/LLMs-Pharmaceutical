```python
from transformers import Trainer, TrainingArguments, AutoModelForSequenceClassification, AutoTokenizer
from datasets import load_dataset
import torch

# Load the IMDb dataset
dataset = load_dataset("imdb")

# Display dataset information
print(dataset)

# Initialize tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModelForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)  # Binary classification (positive/negative)

# Tokenize the dataset
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

tokenized_dataset = dataset.map(tokenize_function, batched=True)

# Split dataset into training and test sets
train_dataset = tokenized_dataset["train"]
eval_dataset = tokenized_dataset["test"]

# Define your training arguments
training_args = TrainingArguments(
    output_dir="./imdb_results",
    num_train_epochs=3,
    per_device_train_batch_size=20,
    per_device_eval_batch_size=20,
    logging_dir="./imdb_blogs",
)

# Define Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
)

# Train the model
trainer.train()
```

```python
from transformers import Trainer, TrainingArguments, AutoModelForSequenceClassification, AutoTokenizer
from datasets import load_dataset
import torch

# Load the CoLA dataset
dataset = load_dataset("glue", "cola")

# Display dataset information
print(dataset)

# Tokenize the dataset
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModelForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)  # Binary classification (positive/negative)

def tokenize_function(examples):
    return tokenizer(examples["sentence"], padding="max_length", truncation=True)

tokenized_dataset = dataset.map(tokenize_function, batched=True)

# Split dataset into training and test sets
train_dataset = tokenized_dataset["train"]
eval_dataset = tokenized_dataset["validation"]

# Define your training arguments
training_args = TrainingArguments(
    output_dir="./gluecola_results",
    num_train_epochs=3,
    per_device_train_batch_size=20,
    per_device_eval_batch_size=20,
    logging_dir="./gluecola_logs",
)

# Define Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
)

# Train the model
trainer.train()
```
