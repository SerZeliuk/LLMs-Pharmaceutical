```python
# Script 1: Trend Analysis for Cancer Drug Development Costs

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data from Sullivan R. et al. and Yuen S. et al.
years = [2000, 2014]
development_costs = [800000000, 2600000000]  # In USD

df = pd.DataFrame({
    'Year': years,
    'Development_Cost': development_costs
})

# Create trend plot
plt.figure(figsize=(10, 6))
sns.regplot(x='Year', y='Development_Cost', data=df, scatter_kws={'s': 100})
plt.title('Cancer Drug Development Cost Trend (2000-2014)')
plt.ylabel('Development Cost (USD)')
plt.axhline(y=0, color='black', linewidth=0.5)

# Format y-axis in billions
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1e9:.1f}B'))

plt.show()
```

```python
# Script 2: Anomaly Detection in Drug ROI

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Data from Sullivan R. et al.
drug_names = ['Average ROI', 'Erlotinib', 'Trastuzumab', 'Rituximab', 'Bevacizumab']
roi_values = [493, 2794, 3421, 2523, 3200]  # ROI in percentage

df = pd.DataFrame({
    'Drug': drug_names,
    'ROI': roi_values
})

# Calculate z-scores for anomaly detection
df['z_score'] = stats.zscore(df['ROI'])
df['is_anomaly'] = abs(df['z_score']) > 2

# Create anomaly detection plot
plt.figure(figsize=(12, 6))
plt.bar(df['Drug'], df['ROI'], color=['red' if x else 'blue' for x in df['is_anomaly']])
plt.title('ROI Analysis with Anomaly Detection')
plt.ylabel('Return on Investment (%)')
plt.xticks(rotation=45)

for i, v in enumerate(df['ROI']):
    plt.text(i, v, f'{v}%', ha='center', va='bottom')

plt.show()
```

```python
# Script 3: Generic Drug Savings Forecast

import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Data from Lee S.M. et al.
periods = ['2014-2018', '2019-2023']
savings = [72000000000, 106000000000]  # In USD

# Create forecast for next period
df = pd.DataFrame({
    'Period': periods,
    'Savings': savings
})

# Simple linear extrapolation for 2024-2028
next_savings = savings[-1] + (savings[-1] - savings[0])
extended_periods = periods + ['2024-2028']
extended_savings = savings + [next_savings]

plt.figure(figsize=(10, 6))
plt.plot(range(len(extended_periods)), extended_savings, 'bo-')
plt.plot(len(extended_periods)-1, extended_savings[-1], 'ro', label='Forecast')

plt.title('Generic Drug Savings Forecast')
plt.xlabel('Time Period')
plt.ylabel('Savings (USD)')
plt.xticks(range(len(extended_periods)), extended_periods, rotation=45)

# Format y-axis in billions
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1e9:.1f}B'))

plt.legend()
plt.grid(True)
plt.show()
```
